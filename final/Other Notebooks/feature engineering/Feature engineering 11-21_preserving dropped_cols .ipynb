```python
#curr_dir=r'C:/Users/megha/Desktop/anaconda codes/masters/sem1/cis520_Machine_Learning/project/pre/Xset/'
comments = pd.read_csv('gdrive/My Drive/520_proj/preFinal/Final/CIS_520_PROJECT_DATA/subSamplesTrain.csv')
comments.head()
```

```python
#Recommendations/count by usertitle

# print(comments.head())
recoSum=comments[comments.userTitle.notnull()][['recommendations','userTitle']].groupby(['userTitle']).sum()
k=comments.groupby(['userTitle']).count().recommendations.to_list()
recoSum['userTitleCount']=k
recoSum['ratioUT']=recoSum['recommendations']/recoSum['userTitleCount']

recoSum = recoSum.reset_index()
recoSum = recoSum.drop(['recommendations','userTitleCount'],axis=1)
recoSum.head()
```

```python
comments = comments.reset_index().merge(recoSum, how="left", on='userTitle').set_index('index')
comments.head()
```

```python
comments.shape
```

```python
comments['editorsSelection']=comments['editorsSelection']*1
```

```python
recoSumUID=comments[comments.userID.notnull()][['recommendations','userID']].groupby(['userID']).sum()
k=comments.groupby(['userID']).count().recommendations.to_list()
recoSumUID['userIDCount']=k
recoSumUID['ratioUID']=recoSumUID['recommendations']/recoSumUID['userIDCount']
# print(recoSumUID.head())
recoSumUID = recoSumUID.reset_index()
recoSumUID = recoSumUID.drop(['recommendations','userIDCount'],axis=1)
#comments_2 = pd.merge(recoSumUID,comments,left_on= 'userID',right_on= 'userID',how = "outer")
recoSumUID.head()
```

```python
comments = comments.reset_index().merge(recoSumUID, how="left", on='userID').set_index('index')
comments.head()
```

```python
# Time of the comments
date_min=pd.DataFrame(comments[['createDate','articleID']].groupby(['articleID'])['createDate'].min()).reset_index()
date_min.rename(columns={'createDate':'min_date'}, inplace=True)
merged = pd.merge(date_min,comments, how='right',on='articleID')
merged['createDate']=merged['createDate']-merged['min_date']
#merged = merged.drop(['min_date'],axis=1)
merged.head()
```

```python
merged.shape
```

```python
featuresFrame = merged[['commentID','articleID','createDate','articleWordCount','depth','recommendations','replyCount','min_date']]
featuresFrame['userTitleRC'] = comments['ratioUT']
featuresFrame['userTitleRC'].fillna(0,inplace=True)
featuresFrame['userIDRC'] = comments['ratioUID']
featuresFrame['userIDRC'].fillna(0,inplace=True)
featuresFrame['editorsSelection']=comments['editorsSelection']
featuresFrame.head()
```

```python
comments.shape,featuresFrame.shape
```

```python
articles['byline']=articles['byline'].apply(lambda x: x.strip())
articles['byline']=articles['byline'].apply(lambda x: x.lower())
articles.head(2)
```

```python
author_art=pd.merge(articles[['articleID','byline']],comments, on='articleID', how='right')
author_art.head(2)
```

```python
A=articles['byline'].value_counts().to_frame().reset_index().rename(columns={'index':'byline','byline':'count_total_article'})
B=author_art['byline'].value_counts().to_frame().reset_index().rename(columns={'index':'byline','byline':'count_total_comment'})
AB=pd.merge(A,B, on='byline', how='inner')
AB['ratio_comment_article']=AB['count_total_comment']/AB['count_total_article']
author_art=pd.merge(author_art,AB[['ratio_comment_article','byline']], on='byline', how='left')
```
