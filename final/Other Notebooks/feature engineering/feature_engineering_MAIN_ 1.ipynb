```python
featuresFrame = merged[['commentID','articleID','createDate','articleWordCount','depth','recommendations','replyCount']]
featuresFrame['userTitleRC'] = comments['ratioUT']
featuresFrame['userTitleRC'].fillna(0,inplace=True)
featuresFrame['userIDRC'] = comments['ratioUID']
featuresFrame['userIDRC'].fillna(0,inplace=True)
featuresFrame['editorsSelection']=comments['editorsSelection']
featuresFrame.head()
```

```python
articles['byline']=articles['byline'].apply(lambda x: x.strip())
articles['byline']=articles['byline'].apply(lambda x: x.lower())
articles.head(2)
```

```python
author_art=pd.merge(articles[['articleID','byline']],comments, on='articleID', how='inner')
author_art.head()
```

```python
A=articles['byline'].value_counts().to_frame().reset_index().rename(columns={'index':'byline','byline':'count_total_article'})
B=author_art['byline'].value_counts().to_frame().reset_index().rename(columns={'index':'byline','byline':'count_total_comment'})
AB=pd.merge(A,B, on='byline', how='inner')
AB['ratio_comment_article']=AB['count_total_comment']/AB['count_total_article']
author_art=pd.merge(author_art,AB[['ratio_comment_article','byline']], on='byline', how='inner')
```

```python
#author_art['commentBody'].str.split(" ")

author_art['word_count']=[ len(i) for i in author_art['commentBody']]
```

```python
author_art.head()
```

```python
ohe=pd.get_dummies(author_art['commentType'])
author_art=pd.merge(author_art, ohe,  left_index=True, right_index=True)
```

```python
author_art_2=author_art[['articleID','commentID','ratioUT','ratioUID','ratio_comment_article','word_count','comment','reporterReply','userReply']]
```

```python
featuresFrame.shape,author_art_2.shape
```

```python
final_features=featuresFrame.merge(author_art_2, on=['articleID','commentID'])
```
